# Synthetic Forecasted Data

## Get Original Data

```python
import pandas as pd
```

```python
df_fc_orig = pd.read_csv('../data/biomass_forecasted.csv')
df_fc_orig.head()
```

```python
d_matrix_orig = pd.read_csv('../../data/Distance_Matrix.csv')
d_matrix_orig.head()
```

## Create Synthetic Data

```python
df_fc_orig['Date'] = pd.to_datetime(df_fc_orig['Date'])
df_fc_orig['year'] = df_fc_orig['Date'].dt.year
df_fc_orig = df_fc_orig[["Latitude", "Longitude", "Forecast", "year"]]
df_fc_proc = df_fc_orig.pivot_table(index=["Latitude", "Longitude"], columns="year", values="Forecast").reset_index()
df_fc_proc.columns = ["Latitude", "Longitude", "2018", "2019"]
df_fc_proc = df_fc_proc.reset_index(names="Index")
```

```python
df_fc_proc
```

# Export Data

```python
df_fc_proc.to_csv('../data/Biomass_History_Forecast.csv', index=False)
d_matrix_orig.to_csv('../data/Distance_Matrix_Synthetic.csv', index=False)
```
