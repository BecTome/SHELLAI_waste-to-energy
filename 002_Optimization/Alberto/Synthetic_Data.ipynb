# Synthetic Forecasted Data

## Get Original Data

```python
import pandas as pd
```

```python
df_fc_orig = pd.read_csv('../../data/Biomass_History.csv')
df_fc_orig.head()
```

```python
d_matrix_orig = pd.read_csv('../../data/Distance_Matrix.csv')
d_matrix_orig.head()
```

## Create Synthetic Data

```python
ls_cols = ['Index', 'Latitude', 'Longitude', '2016', '2017']
df_fc_synthetic = df_fc_orig[ls_cols].copy()
df_fc_synthetic.rename(columns={'2016': '2018', '2017': '2019'}, inplace=True)
df_fc_synthetic.head()
```

# Export Data

```python
df_fc_synthetic.to_csv('../data/Biomass_History_Synthetic.csv', index=False)
d_matrix_orig.to_csv('../data/Distance_Matrix_Synthetic.csv', index=False)
```
